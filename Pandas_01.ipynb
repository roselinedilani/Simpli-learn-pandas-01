```python
df_faa_ai_dataset = pd.DataFrame(df_faa_ai)
```

```python
df_analyze_dataset = df_faa_ai_dataset[['ACFT_MAKE_NAME','LOC_STATE_NAME','ACFT_MODEL_NAME','RMK_TEXT','FLT_PHASE','EVENT_TYPE_DESC','FATAL_FLAG']]
df_analyze_dataset.head(5)
```

```python
df_analyze_dataset['FATAL_FLAG'].fillna(value = 'No',inplace = True)
```

```python
df_analyze_dataset.head()
```

```python
df_analyze_dataset.shape
```

```python
df_finalize_dataset = df_analyze_dataset.dropna(subset=['ACFT_MAKE_NAME'])
```

```python
df_finalize_dataset.shape
```

```python
aircraft_types = df_finalize_dataset.groupby('ACFT_MAKE_NAME')
```

```python
aircraft_types.size()
```

```python
fatal_accidents = df_finalize_dataset.groupby('FATAL_FLAG')
```
